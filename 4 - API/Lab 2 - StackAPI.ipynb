```python
import pandas as pd
import requests
from stackapi import StackAPI
from datetime import datetime
```

#### Question 1: Find the questions and answers of last month.

```python
website = StackAPI('stackoverflow')
from_date = int(datetime.timestamp(datetime(2020,1,1)))
until_date = int(datetime.timestamp(datetime(2020,1,31)))
questions = website.fetch('questions', fromdate=from_date, todate=until_date)
```

```python
pd.json_normalize(questions['items'])
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
from_date = int(datetime.timestamp(datetime(2020,3,21)))
until_date = int(datetime.timestamp(datetime(2020,3,22)))
```

```python
questions = website.fetch('questions', fromdate=from_date, todate=until_date, tagged='python', min=5, sort='votes')
```

```python
question_python = pd.json_normalize(questions['items'])
question_python.max()
```
